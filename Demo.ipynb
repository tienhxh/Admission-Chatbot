```python
from transformers import AutoConfig, AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationalRetrievalChain
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# 1. Tải PhoGPT
model_path = "vinai/PhoGPT-4B-Chat"

config = AutoConfig.from_pretrained(model_path, trust_remote_code=True)
config.init_device = "cuda"
model = AutoModelForCausalLM.from_pretrained(model_path, config=config, load_in_8bit=True)
tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)

# 2. Tạo pipeline PhoGPT cho LangChain
hf_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)
llm = HuggingFacePipeline(pipeline=hf_pipeline)

# 3. Tạo retriever
retriever = vectordb.as_retriever(
    search_kwargs={
        "k": 5,  # Số lượng kết quả trả về
    }
)

# 4. Tạo bộ nhớ hội thoại
memory = ConversationBufferMemory(
    llm=llm,
    output_key="answer",
    memory_key="chat_history",
    return_messages=True
)

# 5. Tạo ConversationalRetrievalChain
chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    chain_type="map_reduce",
    memory=memory,
    verbose=True
)
```

```python
import gradio as gr
import time

# Hàm xử lý câu hỏi
def chat_with_bot(user_input, chat_history=[]):
    try:
        # Giả sử response từ model
        assistant_response = f"Trả lời cho câu hỏi: {user_input}"

        # Cập nhật lịch sử trò chuyện theo định dạng Gradio yêu cầu
        chat_history.append({"role": "user", "content": user_input})
        chat_history.append({"role": "assistant", "content": assistant_response})

        return chat_history, chat_history
    except Exception as e:
        print("Lỗi:", str(e))
        return [{"role": "assistant", "content": f"Lỗi: {str(e)}"}], chat_history

# Tạo giao diện Gradio với các thành phần đẹp và thân thiện
iface = gr.Interface(
    fn=chat_with_bot,
    inputs=[
        gr.Textbox(label="Nhập câu hỏi", placeholder="Hãy nhập câu hỏi của bạn ở đây...", lines=2),
        gr.State()
    ],
    outputs=[
        gr.Chatbot(label="Chatbot Tư vấn Tuyển sinh", type="messages"),
        gr.State()
    ],
    title="Chatbot Tư vấn Tuyển sinh",
    description="Chào mừng đến với chatbot tuyển sinh. Hãy hỏi bất kỳ câu hỏi nào về quy trình tuyển sinh!",
    theme="default",  # Sử dụng theme mặc định
    css="""
        .gradio-container {
            background-image: url('https://vnn-imgs-f.vgcloud.vn/2021/08/11/16/bach-khoa-4.jpg');
            background-size: cover;  /* Đảm bảo ảnh sẽ phủ hết nền */
            background-position: center center;
            background-attachment: fixed;  /* Giữ ảnh nền cố định khi cuộn trang */
        }
        .gradio-input {background-color: #ffffff; border-radius: 10px; padding: 10px;}
        .gradio-output {background-color: #ffffff; border-radius: 10px; padding: 10px;}
        .gradio-button {background-color: #4CAF50; color: white; border-radius: 5px; padding: 10px;}
        .gradio-button:hover {background-color: #45a049;} /* Hover effect cho button */
        .gr-chatbot {border-radius: 10px; padding: 15px; background-color: #ffffff;}
    """,  # CSS tùy chỉnh
    article="Hãy nhập câu hỏi của bạn và chatbot sẽ cung cấp câu trả lời liên quan đến quy trình tuyển sinh!",
)

# Chạy giao diện
iface.launch(share=True, debug=True)
```
